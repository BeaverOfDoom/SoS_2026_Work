## Frekvensanalys
### Förberedelseuppgift
Vid analys av signaler är frekvensinnehållet i signalen ofta av intresse. Fouriertransformen för en tidsdiskret signal (DTFT) ges av:

$X(f) = \sum_{n=-\infty}^{\infty} x(n)\, e^{-j 2 \pi f n}$

$X(f)$ delar upp den tidsdiskreta signalen $x(n)$ i dess frekvenskomponenter och kan därför användas för att analysera frekvensinnehållet i $x(n)$. 
Notera att $X(f)$ är en kontinuerlig funktion av den normerade frekvensen $f$. I numeriska beräkningar används den diskreta Fouriertransformen (DFT), där $X(k)$ beräknas i $N$ punkter $k=0,…,N-1$.

$X(k) = \sum_{n=-\infty}^{\infty} x(n)\, e^{\left(-j \tfrac{2\pi k n}{N}\right)}$

I praktiken hanterar vi vanligtvis signaler som är ändligt långa, och vi antar då att $x(n)=0$ för $n<0$ och $n>M-1$.  

$X(k) = \sum_{n=0}^{M-1} x(n)\, e^{\left(-j \tfrac{2\pi k n}{M}\right)}$



Fast Fourier Transform (FFT) är en beräkingseffektiv implementation av DFT. Funktionen *np.fft.fft(x,N)* i Pyton beräknar *N*-punkts DFT:n av *x*. Om det andra argumentet utelämnas, sätts *N* till längden av *x*.

Notera att $X(f)$ och $X(k)$ är komplexa tal. Vid signalanalys är vi vanligtvis mest intresserade av hur mycket av signalens energi som finns vid olika frekvenser. Därför är vi ofta intresserade av magnitudspektrat $|X(f)|$ eller effektspektrat $|X(f)|^2$.

## FFT av Cosinus signal
### Uppgift 2.1

Skapa en tidsdiskret cosinussignal med frekvens $f=0.25$ och en tidsdiskret cosinussignal med frekvens $f=0.3$ av längd $N=1000$. Skapa också en ny tidsdiskret signal genom att summera cosinussignalerna.
Beräkna och plotta 1000-punkters DFT för de tre signalerna.

```python

# Parametrar
f1 = 0.25
f2 = 0.3
M = 1000
n = np.arange(1, M+1)  # 1 till M

# Tidsdiskreta cosinussignaler
x1 = np.cos(2 * np.pi * f1 * n)
x2 = np.cos(2 * np.pi * f2 * n)
x3 = x1 + x2

# Beräkna N-punkts DFT
N = 1000
X1 = np.fft.fft(x1, N)
X2 = np.fft.fft(x2, N)
X3 = np.fft.fft(x3, N)

# Plot DFT som funktion av index k
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

ax1.plot(np.abs(X1))
ax1.set_ylabel('|X1(k)|')
ax1.grid()

ax2.plot(np.abs(X2))
ax2.set_ylabel('|X2(k)|')
ax2.grid()

ax3.plot(np.abs(X3))
ax3.set_ylabel('|X3(k)|')
ax3.set_xlabel('k')
ax3.grid()

plt.tight_layout()
plt.show()
```

### Uppgift 4.3 - STFT
I ett elektroencefalogram (EEG) registreras hjärnbarkens spontana elektriska aktivitet med hjälp av elektroder på skalpen. Filen *eeg.npz* består av 26-kanals EEG inspelat före, under och efter ett kort epileptiskt anfall. Samplingsfrekvensen är $Fs=256$ Hz.
Läs in filen och plotta kanal 1-3.
Hur lång är inspelningen?
När tror du att epelipsianfallet inträffar?

```python
# Ladda filen
data = np.load('eeg.npz')

# Skriv ut alla variabelnamn
print("Variabler i filen:", list(data.keys()))

eeg = data['X']
print("eeg storlek:", eeg.shape)

# Definiera samplings frekvens (Fs)
fs_eeg = 256 

# Skapa tidsvektor
t = np.arange(0, eeg.shape[0] / fs_eeg, 1 / fs_eeg)

# Skapa figur med 3 subplots som delar x-axel
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

#Välj ut kanal 1-3 och plotta
x1=eeg[:,0]
x2=eeg[:,1]
x3=eeg[:,2]

ax1.plot(t,x1)
ax1.set_ylabel('x1')
ax1.set_title('EEG')
ax1.grid()
ax2.plot(t,x2)
ax2.set_ylabel('x2')
ax2.grid()
ax3.plot(t,x3)
ax3.set_ylabel('x3')
ax3.set_xlabel('t')
ax3.grid()
```
