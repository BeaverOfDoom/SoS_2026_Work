## Förberedelse
Ett linjärt tidsinvariant (LTI) system beskrivs fullständigt av sin överföringsfunktion 
$H(z) = \frac{B(z)}{A(z)} 
     = G \, \frac{\prod_{k=1}^{M} \left(1 - n_k z^{-1}\right)}
                 {\prod_{k=1}^{N} \left(1 - p_k z^{-1}\right)}$
                 
$n1, n2,…nM$ betecknar systemets M nollställen

$p1,p2,..,pN$ betecknar systemets N poler

$G$ är en konstant förstäkningsfaktor (gain)

Frekvenssvaret hos systemet fås genom att avläsa H(z) på enhetscirkeln
$H(\omega) = H(z)\big|_{z = e^{j\omega}} 
= G \, \frac{\prod_{k=1}^{M} \left(1 - n_k e^{-j\omega}\right)}
             {\prod_{k=1}^{N} \left(1 - p_k e^{-j\omega}\right)}$

Frekvenssvaret kan uttryckas som funktion av normerad frekvens $f$ istället för som funktion av normerad vinkelfrekvens, där $\omega=2 \pi f$. 

Förstärkning vid frekvens $f$ ges av frekvenssvarets magnitud $|H(f)|$. Frekvenssvaret kan därför enkelt skattas från systemets pol-nollställe plot; $|H(f)|$ blir stor nära en pol och liten nära ett nollställe.

```python

# Lista för nollställen och poler
zeros = []
poles = []

dragging = None  # Håller koll på vilken markör som dras
drag_type = None  # 'zero' eller 'pole'


# --- Setup figurer med GridSpec ---
#fig = plt.figure(figsize=(12,6))

fig = plt.figure(figsize=(12,12))

gs = GridSpec(2, 2, figure=fig, width_ratios=[1, 1.2])

ax_z = fig.add_subplot(gs[:, 0])  # Z-planen tar hela vänsterkolumnen
ax_freq = fig.add_subplot(gs[0, 1])  # Frekvenssvar överst höger
ax_imp = fig.add_subplot(gs[1, 1])   # Impulssvar nederst höger


# --- Z-plan ---
ax_z.set_title("Pol-nollställe diagram")
ax_z.set_xlim([-1.5, 1.5])
ax_z.set_ylim([-1.5, 1.5])
ax_z.set_aspect('equal', adjustable='box')
ax_z.grid(True)

# Rita enhetscirkeln
circle = plt.Circle((0,0), 1, color='black', fill=False, linestyle="--")
ax_z.add_artist(circle)

zero_scat = ax_z.scatter([], [], marker='o', facecolors='none', edgecolors='b', s=100, picker=5, label='Nollställen')
pole_scat = ax_z.scatter([], [], marker='x', color='r', s=100, picker=5, label='Poler')
ax_z.legend(loc='upper right')


# Reset-knapp
reset_ax = fig.add_axes([0.05, 0.02, 0.1, 0.05])
reset_button = Button(reset_ax, 'Reset')


# --- Frekvenssvar ---
ax_freq.set_title("Frekvenssvar")
ax_freq.set_xlabel("Normerad frekvens f")
ax_freq.set_ylabel("Amplitude / Fas [rad]")
ax_freq.grid(True)
amp_line, = ax_freq.plot([], [], label='Amplitude')
phase_line, = ax_freq.plot([], [], label='Fas')
ax_freq.legend()

# --- Impulssvar ---
ax_imp.set_title("Impulssvar")
ax_imp.set_xlabel("n")
ax_imp.set_ylabel("Amplitud")
ax_imp.grid(True)

def update_plots():
    """Uppdatera alla plottar"""
    # Scatter-uppdatering
    zero_scat.set_offsets(np.c_[np.real(zeros), np.imag(zeros)] if zeros else np.empty((0,2)))
    pole_scat.set_offsets(np.c_[np.real(poles), np.imag(poles)] if poles else np.empty((0,2)))
    
    # Överföringsfunktion
    b, a = signal.zpk2tf(zeros, poles, 1)
    
    # Frekvenssvar
    w, h = signal.freqz(b, a, worN=512, whole=True)
    amp_line.set_data(w, np.abs(h))
    phase_line.set_data(w, np.angle(h))
    ax_freq.relim()
    ax_freq.autoscale_view()
    ax_freq.set_xlim([0, 2*np.pi])
    ax_freq.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax_freq.set_xticklabels(['0', '0.25', '0.5', '0.75', '1'])
    
    # Impulssvar
    imp = np.zeros(50)
    imp[0] = 1
    resp = signal.lfilter(b, a, imp)
    ax_imp.cla()
    ax_imp.set_title("Impulssvar")
    ax_imp.set_xlabel("n")
    ax_imp.set_ylabel("Amplitud")
    ax_imp.grid(True)
    ax_imp.stem(np.arange(len(resp)), resp.real, basefmt=" ")

    fig.canvas.draw_idle()


def onclick(event):
    global dragging, drag_type
    if event.inaxes != ax_z:
        return
    z = event.xdata + 1j*event.ydata
    tol = 1e-1

    if event.dblclick:  # Dubbelklick: ta bort närmaste punkt
        remove_point(z, tol)
        return

    if event.button == 1 and not event.key == 'control':  # vänsterklick + inte Ctrl = nollställe
        if any(abs(z - z0) < tol for z0 in zeros):
            return
        if abs(z.imag) < tol:
            zeros.append(np.real(z))
        else:
            zeros.extend([z, np.conjugate(z)])

    elif event.button == 3 or (event.button == 1 and event.key == 'control'):  # högerklick eller Ctrl+klick = pol
        if any(abs(z - p0) < tol for p0 in poles):
            return
        if abs(z.imag) < tol:
            poles.append(np.real(z))
        else:
            poles.extend([z, np.conjugate(z)])

    update_plots()

def remove_point(z, tol):
    """Ta bort närmast pol eller noll om inom tolerans."""
    global zeros, poles
    for lst in (zeros, poles):
        for val in lst:
            if abs(z - val) < tol:
                lst.remove(val)
                if val.imag != 0 and np.conjugate(val) in lst:
                    lst.remove(np.conjugate(val))
                update_plots()
                return

def onpick(event):
    global dragging, drag_type
    ind = event.ind[0]

    if event.artist == zero_scat:
        dragging = ind
        drag_type = 'zero'
    elif event.artist == pole_scat:
        dragging = ind
        drag_type = 'pole'

def onmotion(event):
    global dragging
    if dragging is None or event.inaxes != ax_z:
        return
    z_new = event.xdata + 1j*event.ydata
    tol = 1e-6

    if drag_type == 'zero':
        z_old = zeros[dragging]
        zeros[dragging] = z_new
        for i, z in enumerate(zeros):
            if i != dragging and abs(z - np.conjugate(z_old)) < tol:
                zeros[i] = np.conjugate(z_new)
                break

    elif drag_type == 'pole':
        p_old = poles[dragging]
        poles[dragging] = z_new
        for i, p in enumerate(poles):
            if i != dragging and abs(p - np.conjugate(p_old)) < tol:
                poles[i] = np.conjugate(z_new)
                break

    update_plots()

def onrelease(event):
    global dragging, drag_type
    dragging = None
    drag_type = None

def reset(event):
    """Nollställ alla poler och nollställen."""
    global zeros, poles
    zeros.clear()
    poles.clear()
    update_plots()

# Koppla events
fig.canvas.mpl_connect('button_press_event', onclick)
fig.canvas.mpl_connect('pick_event', onpick)
fig.canvas.mpl_connect('motion_notify_event', onmotion)
fig.canvas.mpl_connect('button_release_event', onrelease)
reset_button.on_clicked(reset)

# --- Viktigt: initiera alla plottar vid start ---
update_plots()
fig.subplots_adjust(hspace=0.5, top=0.95, bottom=0.1)
plt.show()
```
