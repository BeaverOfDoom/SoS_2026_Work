## 1. Förberedelse
Ett linjärt tidsinvariant system beskrivs av följande differensekvation:

$y(n) = -\sum_{k=1}^{N} a_k\, y(n-k) + \sum_{k=0}^{M} b_k\, x(n-k)$

I Pyton finns funktionen lfilter, som implementerar denna operation *y=lfilter(b,a,x)*, 
där 

*b* är en vektor med koefficienterna $b_0, b_1, …, b_M$ 

*a* är en vektor med koefficienterna $a_0, a_1, …, a_N$, där $a_0=1$ är skalningskoefficienten framför $y(n)$ i differensekvationen.

*x* är insignalen

*y* är utsignalen

## 6. Impulssvar
Ett filters impulssvar h(n) ges av utsignalen då insignalen är en enhetsimpuls, dvs då
 
$
x(n) = \delta(n) =
\begin{cases}
1, & n = 0 \\
0, & n \neq 0
\end{cases}
$

### Uppgift 6.1
Simulera en enhetsimpuls genom att skapa en signal med en etta följd av 20 nollor och filtrera den med respektive filter från uppgift 2 till 5.
Vilka skillnader och likheter ser du mellan impulssvaren?


```python
# Skapa enhetsimpuls: en etta följd av 20 nollor
x_imp = np.concatenate(([1], np.zeros(20)))


# Filtrera signalen x_imp genom respektive filter (FIR LP, FIR HP, IIR LP, IIR HP) 
h_FIR_LP = lfilter(b_FIR_LP, a_FIR, x_imp)
h_FIR_HP = lfilter(b_FIR_HP, a_FIR, x_imp)
h_IIR_LP = lfilter(b_IIR_HP, a_IIR_LP, x_imp)
h_IIR_HP = lfilter(b_IIR_HP, a_IIR_HP, x_imp)

# Plotta impuls och impulssvar
fig, axes = plt.subplots(5, 1, figsize=(10, 10), sharex=True)

axes[0].stem(x_imp)
axes[0].set_title("Insignal (enhetsimpuls)")
axes[0].grid()


axes[1].stem(h_FIR_LP)
axes[1].set_title("Impulssvar FIR LP")
axes[1].grid()

axes[2].stem(h_FIR_HP)
axes[2].set_title("Impulssvar FIR HP")
axes[2].grid()

axes[3].stem(h_IIR_LP)
axes[3].set_title("Impulssvar IIR LP")
axes[3].grid()

axes[4].stem(h_IIR_HP)
axes[4].set_title("Impulssvar IIR HP")
axes[4].set_xlabel("n")
axes[4].grid()

plt.tight_layout()
plt.show()
```

### Uppgift 6.2
Utsignalen från ett linjärt tidsinvariant system fås genom att falta dess impulssvar med insignalen. 

$y(n) = x(n) * h(n) = \sum_{k=-\infty}^{\infty} x(k) \, h(n-k)$

I NumPy finns operatioonen *convolve* som implementerar denna operation. 

Falta signalen $x_3$ som användes i uppgift 2-5 med respektive impulssvar från uppgift 6. 
Vad ser du? Stämmer resultatet med dina förväntningar? 


```python
f2 = 0.01
N = 200
n = np.arange(0, N+1)
x2 = np.cos(2 * np.pi * f2 * n)


x3 = x1 + x2

#Faltning
y3_conv_h_FIR_LP = np.convolve(x3, h_FIR_LP)
y3_conv_h_FIR_HP = np.convolve(x3, h_FIR_HP)
y3_conv_h_IIR_LP = np.convolve(x3, h_IIR_LP)
y3_conv_h_IIR_HP = np.convolve(x3, h_IIR_HP)

# Plotta resultaten
fig, axes  = plt.subplots(5, 1, figsize=(10, 8), sharex=True)


axes[0].plot(x3)
axes[0].set_title('Insignal')
axes[0].grid()

axes[1].plot(y3_FIR_LP, label='y3_FIR_LP')
axes[1].plot(y3_conv_h_FIR_LP, 'r:', label='conv med h_FIR_LP')
axes[1].set_title('FIR LP filtrerad')
axes[1].legend()
axes[1].grid()

axes[2].plot(y3_FIR_HP, label='y3_FIR_HP')
axes[2].plot(y3_conv_h_FIR_HP, 'r:', label='conv med h_FIR_HP')
axes[2].set_title('FIR HP filtrerad')
axes[2].legend()
axes[2].grid()

axes[3].plot(y3_IIR_LP, label='y3_IIR_LP')
axes[3].plot(y3_conv_h_IIR_LP, 'r:', label='conv med h_IIR_LP')
axes[3].set_title('IIR LP filtrerad')
axes[3].legend()
axes[3].grid()

axes[4].plot(y3_IIR_HP, label='y3_IIR_HP')
axes[4].plot(y3_conv_h_IIR_HP, 'r:', label='conv med h_IIR_HP')
axes[4].set_title('IIR HP filtrerad')
axes[4].legend()
axes[4].grid()

plt.tight_layout()
plt.show()
```
